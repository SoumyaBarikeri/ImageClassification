```python
# First, you should install flickrapi
# pip install flickrapi

import flickrapi
import urllib.request
from PIL import Image
```

```python
# Flickr api access key 
flickr=flickrapi.FlickrAPI('c6a2c45591d4973ff525042472446ca2', '202ffe6f387ce29b', cache=True)


keyword = 'Tiger'

photos = flickr.walk(text=keyword,
                     tag_mode='all',
                     tags=keyword,
                     extras='url_c',
                     per_page=100,           # may be you can try different numbers..
                     sort='relevance')
```

```python
urls = []
for i, photo in enumerate(photos):
    print (i)
    
    url = photo.get('url_c')
    urls.append(url)
    
    # get 50 urls
    if i > 5000:
        break

print (urls)
```

```python
valid_urls = []
i = 1

for new_url in urls:    
    if(new_url != None):
        valid_urls.append(new_url)
        i = i+1
        
#         if i > 2000:
#             break

print(valid_urls)
```

```python
# Download image from the url and save it to '00001.jpg'
j=1
for url in valid_urls:
    urllib.request.urlretrieve(url, "Tiger/im_{}.jpg".format(j))
    j=j+1
```

```python
# Resize the image and overwrite it
for l in range(j)
    image = Image.open('Tiger/im_%l.jpg') 
    image = image.resize((255, 255), Image.ANTIALIAS)
    image.save('Tiger/im_%l.jpg')
```
